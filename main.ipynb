```python
import cv2
import numpy as np
from tensorflow.keras.models import load_model
```

```python
model = load_model("my_model.h5")
```

```python
cap = cv2.VideoCapture(0)
```

```python
while True:
    ret, frame = cap.read()
    
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    gray_resized = cv2.resize(gray, (28, 28), interpolation=cv2.INTER_AREA)
    
    prediction = model.predict(np.expand_dims(gray_resized, axis=0))
    digit = np.argmax(prediction)
    
    cv2.putText(frame, str(digit), (10, 50), cv2.FONT_HERSHEY_SIMPLEX, 2, (0, 255, 0), 2)
    
    cv2.imshow('Digit Recognition', frame)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):q
        break

cap.release()
cv2.destroyAllWindows()
```
